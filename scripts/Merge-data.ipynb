## Load both datasets

```python
import pandas as pd

# File paths
iris_sample_path = "irissample.csv"
iris_extra_path = "iris_.csv"

```

## Merge Both Datasets

```python

iris_df = pd.read_csv(iris_sample_path)
extra_df = pd.read_csv(iris_extra_path)

# Add 'Flower_ID' column to extra_df to match iris_df
extra_df["Flower_ID"] = range(1, len(extra_df) + 1)

# Perform the merge
merged_df = pd.merge(iris_df, extra_df, on="Flower_ID", how="left")

# Save merged dataset
merged_df.to_csv("../results/merged_dataset.csv", index=False)

# Display merged dataset preview
print("Merged dataset preview:")
display(merged_df.head())
```

## Fix Duplicate Column Names

```python
# Drop duplicate columns (keep only '_x' version)
merged_df = merged_df.drop(columns=['Sepal_Length_y', 'Sepal_Width_y', 'Petal_Length_y', 'Petal_Width_y', 'Species_y'])

# Rename remaining columns
merged_df = merged_df.rename(columns={
    'Sepal_Length_x': 'Sepal_Length',
    'Sepal_Width_x': 'Sepal_Width',
    'Petal_Length_x': 'Petal_Length',
    'Petal_Width_x': 'Petal_Width',
    'Species_x': 'Species'
})

# Save cleaned merged dataset
merged_df.to_csv("../results/merged_dataset.csv", index=False)

# Display final cleaned dataset
print("Final Merged Dataset:")
display(merged_df.head())
```
